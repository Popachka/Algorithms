## <center> Введение в деревья. </center>

### Преимущества:
1. **Быстрый поиск**: Поиск занимает среднюю временную сложность O(log n), где n - кол-во узлов в дереве. Это намного быстрее, чем поиск элемента в массиве или связанном списке, которы в худшем случае имеют временную сложность O(n)
2. **Обход по порядку**: В этой структуре можно передвигаться только учитывая строгий порядок, что означает это можно использовать для сортировки набора данных
3. **Экономия места**: Не хранят какой либо избыточной информации
### Недостатки:
1.  **Перекошенность**: В худшем случае delete, insert будут работать за линейное время
2. **Эффективность**: Не эффективны для наборов данных с большим кол-вом дублей, поскольку они будут тратить место.

```python
class Tree:
    def __init__(self, val = None) -> None:
        self.value = val
        if self.value:
            self.left = Tree()
            self.right = Tree()
        else:
            self.left = None
            self.right = None

    def isempty(self):
        return self.value == None
    def isleaf(self):
        if self.left.left == None and self.right.right == None:
            return True
        else:
            False
    def insert(self, data): # O(log n) space: O(n)
        if self.isempty():
            self.value = data
            self.left = Tree()
            self.right = Tree()
            print("{} is inserted successfully".format(self.value))
        elif data < self.value:
            self.left.insert(data)
        elif data > self.value:
            self.right.insert(data)
        elif data == self.value:
            return
    def find(self, v ): # O(log n) space: O(n)
        if self.isempty():
            print("{} is not found".format(v))
            return False
        if self.value == v:
            print("{} is found".format(v))
            return True
        if self.value > v:
            return self.left.find(v)
        else:
            return self.right.find(v)
    def inorder(self):
        if self.isempty():
            return []
        else:
            return self.left.inorder() + [self.value] + self.right.inorder()
    def maxval(self):
        if self.right.right == None:
            return (self.value)
        else:
            return (self.right.maxval())
    def delete(self, v): # O(log n) space: O(1)
        if self.isempty():
            return
        if v < self.value:
            self.left.delete(v)
            return
        if v > self.value:
            self.right.delete(v)
            return
        if v == self.value:
            if self.isleaf():
                self.value = None
                self.right = None
                self.left = None
                return
            elif self.left.isempty():
                self.value = self.right.value
                self.left = self.right.left
                self.right = self.right.right
            else:
                self.value = self.left.maxval()
                self.left.delete(self.left.maxval())
                return
t = Tree(5)
t.insert(10)
t.insert(11)
t.insert(1)
t.insert(1)
t.find(11)
t.find(2)
print(t.inorder())
t.delete(10)
print(t.inorder())
```
